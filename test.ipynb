```python
import os
import cv2
import matplotlib.pyplot as plt
from _paths import nomeroff_net_dir
from glob import glob
import json
import warnings
from metrics.per_symbol import calculate_metrics
warnings.filterwarnings("ignore")

from nomeroff_net import pipeline
from nomeroff_net.tools import unzip
from nomeroff_net.pipes.number_plate_localizators.yolo_v8_detector import Detector
```

```python
number_plate_detection_and_reading = pipeline("number_plate_detection_and_reading", 
                                              number_plate_localization_detector=Detector,
                                              image_loader="opencv")
```

```python
DATASET_PATH_IMG = "/workspace/datasets/Armenia/test/img/"
DATASET_PATH_ANN = "/workspace/datasets/Armenia/test/ann/"

predicted_texts = []
target_texts = []

for js in os.listdir(DATASET_PATH_ANN):
    with open(DATASET_PATH_ANN+js, 'r') as f:
        data = json.load(f)
    target_texts.append(data['description'])


for car in os.listdir(DATASET_PATH_IMG):
    result = number_plate_detection_and_reading(
        glob(os.path.join(nomeroff_net_dir, DATASET_PATH_IMG+car)),
    )
    (images, images_bboxs, images_points, images_zones, region_ids,region_names, count_lines, confidences, texts) = unzip(result)
    predicted_texts.append(texts)
```

```python
for predicted, actual in zip(predicted_texts, target_texts):
    print(f"Предсказанная метка: {predicted}")
    print(f"Настоящая метка: {actual}")
```
